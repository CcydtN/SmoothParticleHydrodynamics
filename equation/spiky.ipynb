```python
from sympy import *
from sympy.physics.mechanics import *
from IPython.display import display
import util
import math
init_vprinting()
```

```python
h, r, pi = symbols("h r π")

func = (h-r)**3

volume = integrate(4*pi*r**2*func, (r, 0, h))
simplify(volume)
```

```python
function, gradient, laplacian = util.get_derivatives(func, r)
display(function, gradient, laplacian)
```

```python
test_h = 2 # don't use 1, casue it can't test case like 1/(h**2)
upper_bound = test_h

s = [(h, test_h), (pi, math.pi), (r, abs(r))]
f = (func/volume).subs(s)
g = (gradient/volume).subs(s)
l = (laplacian/volume).subs(s)

util.sampling(f, g, l, r, upper_bound, test_h, "samples/spiky.json")
```

```python
plot(f,g,l, (r, -upper_bound, upper_bound))
```
